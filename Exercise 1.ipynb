$$\oint_CP(x,y)dx+Q(x,y)dy = \iint_{\mathbb{D}}\frac{\partial Q}{\partial x}-\frac{\partial P}{\partial y} dA$$

We know that the Area of $\mathbb{D}$ is $\iint_\mathbb{D}dA$, so we can use Green's Theorem to calculate the area of $\mathbb{D}$ by selecting $P(x,y)=\frac{-y}{2}$ and $Q(x,y)=\frac{x}{2}$, so that:

$$\iint_{\mathbb{D}}\frac{\partial Q}{\partial x}-\frac{\partial P}{\partial y} dA = \iint_{\mathbb{D}}\frac{1}{2}-\frac{(-1)}{2} dA = \iint_{\mathbb{D}}1.dA=\iint_{\mathbb{D}}dA$$

Let $A$ be the Area of $\mathbb{D}$ then:

$$A = \iint_{\mathbb{D}}dA=\oint_CP(x,y)dx+Q(x,y)dy=\oint_C\frac{-y}{2}dx+\frac{x}{2}dy = \frac{1}{2}\oint_Cxdy-ydx=\frac{1}{2}(\oint_{C_0}xdy-ydx+\oint_{C_1}xdy-ydx+\dots+\oint_{C_n}xdy-ydx)$$

To calculate the $k$-th line integral, we would need to parametrize the segment from $(x_k,y_k)$ to $(x_{k+1},y_{k+1})$.

$$C_k=(x_k+(x_{k+1}-x_k)t,y_k+(y_{k+1}-y_k)t), t\in[0,1]$$

Substituting the parametrization on the integral we get:

$$\oint_{C_k}xdy-ydx=(y_kx_{k+1}-x_ky_{k+1})$$
